```python
name= []
owner= []
price= []
emi = []
location = []
city_list = [2,134,5,2378,2423,4709,3686,5732,777,1692,2987,769,6105,6,132,2130,671,6351,290,8184]
city_name = ["New delhi","Noida","Gurgaon","Mumbai","pune","Bengaluru","Hyderabad","chennai","kolkata","Ahmedabad","Bhopal","Chandigarh","coimbatore","Faridabad","Ghaziabad","jaipur","jallandhar","cochin","lucknow","patna"]
            
for i in city_list:
    url = 'https://www.cars24.com/buy-used-car/?sort=P&storeCityId={}'.format(i)
    page=requests.get(url)
    soup= BeautifulSoup(page.text)
    for val in soup.find_all("h2",class_='_3FpCg'):
        name.append(val.text)
    for j in soup.find_all("ul",class_="bVR0c"):
        owner.append(j.text)
    for x in soup.find_all("div",class_="_7udZZ"):
        price.append(x.text)
    for e in soup.find_all("div",class_="_2HFRN"):
        emi.append(e.text)
        location.append(i)
    
```
